```python
from openai import OpenAI

# Connect to local Ollama server
client = OpenAI(
    base_url="http://localhost:11434/v1",
    api_key="ollama"  
)

def ask_llm(model, prompt):
    response = client.chat.completions.create(
        model=model,
        messages=[{"role": "user", "content": prompt}]
    )
    return response.choices[0].message.content

# Test connection (any model you already have)
print(ask_llm("gemma3:270m", "What is sentiment analysis? Answer in one sentence."))

```

```python
import pandas as pd

df = pd.read_csv("data/emails.csv", sep=";")


from openai import OpenAI

client = OpenAI(
    base_url="http://localhost:11434/v1",
    api_key="ollama"
)

def classify_email(model, headline):
    prompt = f"""
Classify this email headline as one of:
- spam
- work
- unknown

Return ONLY one word.

Headline: {headline}
"""

    response = client.chat.completions.create(
        model=model,
        messages=[{"role": "user", "content": prompt}]
    )

    return response.choices[0].message.content.strip().lower()


```
