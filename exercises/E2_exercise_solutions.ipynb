
### Exercise 7: Selecting with `.loc`

Use the `products` DataFrame from Exercise 5 for this exercise.

a) Select and print the row at index `1` using `.loc`. What type is the result?

b) Select and print the row at index `1` using `.loc` with double brackets. What type is the result now?

c) Select and print only the `product` and `price` columns for all rows using `.loc`.

d) Select and print the `price` and `units_sold` columns for rows 1 to 2 (inclusive) using `.loc`.

````python
# a) Single bracket .loc returns a Series

print(products.loc[1])
print(type(products.loc[1]))
````

````python
# b) Double bracket .loc returns a DataFrame

print(products.loc[[1]])
print(type(products.loc[[1]]))
````

````python
# c) Select product and price columns for all rows

print(products.loc[:, ["product", "price"]])
````

````python
# d) Select price and units_sold for rows 1 to 2 (inclusive)

print(products.loc[1:2, ["price", "units_sold"]])
````

### Exercise 8: Selecting with `.iloc`

Using the same `products` DataFrame:

a) Select the first two rows using `.iloc`.

b) Select the element in the third row, second column using `.iloc`.

c) Select all rows but only the first and third columns using `.iloc`.

d) What is the difference between `df.loc[0:2]` and `df.iloc[0:2]`? Test both and explain in a comment.

````python
# a) First two rows

print(products.iloc[0:2])
````

````python
# b) Element in third row, second column

print(products.iloc[2, 1])
````

````python
# c) All rows, first and third columns

print(products.iloc[:, [0, 2]])
````

````python
# d) Difference between .loc[0:2] and .iloc[0:2]

print(products.loc[0:2])   # Returns rows with labels 0, 1, 2 (inclusive)
print()
print(products.iloc[0:2])  # Returns rows at positions 0, 1 (exclusive upper bound)

# .loc uses label-based indexing where the endpoint is inclusive.
# .iloc uses position-based indexing where the endpoint is exclusive,
# just like standard Python slicing.
````

### Exercise 10: Putting It All Together

A small shop tracks weekly sales in a dictionary:

```python
sales_data = {
    "day": ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"],
    "customers": [30, 45, 50, 35, 60, 80, 70],
    "sales_eur": [600, 900, 1100, 750, 1300, 1800, 1500]
}
```

a) Create a DataFrame from this dictionary.

b) Add a column `avg_spend` that computes the average spending per customer (i.e., `sales_eur / customers`).

c) Using boolean indexing, filter the DataFrame to show only days where the number of customers exceeded 50.

d) Using `.iloc`, select the data for the weekend (last two rows) and all columns except `day`.

e) What was the total revenue for the week? Use NumPy or Pandas to compute and print the answer.

````python
# a) Create the DataFrame

sales_data = {
    "day": ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"],
    "customers": [30, 45, 50, 35, 60, 80, 70],
    "sales_eur": [600, 900, 1100, 750, 1300, 1800, 1500]
}

sales = pd.DataFrame(sales_data)
sales
````

````python
# b) Add avg_spend column

sales["avg_spend"] = sales["sales_eur"] / sales["customers"]
````

````python
# c) Filter days where customers exceeded 50

print(sales[sales["customers"] > 50])
````

````python
# d) Weekend data (last two rows), all columns except day

print(sales.iloc[-2:, 1:])
````

````python
# e) Total revenue for the week

print(f"Total weekly revenue: {sales['sales_eur'].sum()}€")
````
