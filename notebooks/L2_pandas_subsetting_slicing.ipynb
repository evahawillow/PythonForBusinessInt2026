This notebook demonstrates how you can select specific rows and columns from a Pandas dataframe.

```python
import pandas as pd
```

```python
# Loading the UCI Online Retail dataset into this Jupyter Notebook
# First download the dataset from the source: https://archive.ics.uci.edu/ml/datasets/Online+Retail
# Place the dataset in the same folder as this notebook (or specify the path below within "")
# This dataset contains transactional data from a UK-based online retailer (2010-2011).

df = pd.read_excel("Online Retail.xlsx")
```

```python
# What the dataset looks like

df
```

## Using .loc 

### Selecting rows and returning a Pandas series or DataFrame

```python
# Single bracket returns a series

df.loc[1]
```

```python
type(df.loc[1])
```

```python
# Double bracket returns a DataFrame

df.loc[[1]]
```

```python
type(df.loc[[1]])
```

### Slicing

Slicing can be applied to rows or columns with df[a:b] where a is the start and b is the end. Note that both start and end values are included in the result.

```python
# Slicing only based on rows

df.loc[0:5]
```

```python
# Slicing based on columns, maintaining all rows. Note that : means all 

df.loc[:, ["InvoiceNo", "Description"]]
```

Note how the desired columns are given as a list if there are multiple.

```python
# When returning only one column the column name does not have to be given as a list

df.loc[:, "Description"]
```

```python
# Specifying both rows and columns, wanting the unit price and quantity for rows 5-7

df.loc[5:7, ["UnitPrice", "Quantity"]]
```

### Boolean example

We want to get a subset of the dataset with rows where the UnitPrice is above 10. Here is one approach to doing it.

```python
# Getting the boolean array

expensive = df["UnitPrice"] > 10

expensive
```

```python
# Subsetting the dataset with this boolean array

df[expensive]
```

```python
# Note that this could done with just one row

df[df["UnitPrice"] > 10]
```

Elegant? No. But it does what it is supposed to do.

## Using .iloc

### Selecting rows and returning a Pandas series or DataFrame

```python
# Single bracket returns an array

df.loc[1]
```

```python
# Double bracket returns a DataFrame

df.iloc[[1]]
```

```python
# Notice the difference when comparing .loc and iloc. The slice returns only the five first positions. 
# With .loc it would return six rows, while iloc returns five. 

df.iloc[0:5]
```

```python
# Slicing based on column positions, maintaining all rows.


df.iloc[:, 0:2] 
```

## Selecting data without .iloc and .loc

```python
# Selecting rows

df[0:5]
```

```python
# Selecting a column

df["Description"]
```

```python
# Selecting multiple columns

df[["Description", "Country"]]
```

```python
# Selecting rows and columns

df[0:5][["Description", "Country"]]
```

## A note on subsetting DataFrames

```python
# Let's take a subset of the first five rows and store it

first_five = df[0:5]

first_five
```

```python
# Let's change the country to "Narnia" in this subset

first_five.loc[0:5, "Country"] = "Narnia"

first_five
```

```python
# Now let's look at our original dataset

df
```

Oh no! Looks like we inadvertently changed the original dataset. How did this happen? Because both were in fact referring to the same object. The first_five did not create a copy of the first five rows, but rather referenced the five rows inside the object df.

```python
# Restoring the original df

df = pd.read_excel("Online Retail.xlsx")
```

Now we have the dataframe in its original form. This would be a pain if the dataset was massive and took long to load...

How can we avoid this issue?

```python
# Use the .copy() method of pandas to make a true copy rather than referencing the same object

first_five = df.copy()[0:5]
```

```python
first_five.loc[0:5, "Country"] = "Narnia"
```
